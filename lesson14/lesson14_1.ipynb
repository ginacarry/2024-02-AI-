```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

#:sudo apt-get update -y
#:sudo apt-get install -y fonts-wqy-zenhei

#codespace
font = FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=8)
#windows
#font=FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=8)
#============#
#mac
#font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=8)

student1 = {'':78, '': 80, '':75, '': 85, '':92}
student2 = {'':90, '': 75, '':98, '': 65, '':75}

subjects=list(student1.keys())
scores1=list(student1.values())
scores2=list(student2.values())

fig=plt.figure(figsize=(8, 5),dpi=250,facecolor='#cccccc')
axes=fig.add_subplot(1,1,1)
axes.plot(subjects,scores1,'r^-.',label='1')
axes.plot(subjects,scores2,'go--',label='2')
axes.set_xticklabels(subjects,fontproperties=font)
axes.set_title("12",fontproperties=font)
axes.set_xlabel("",fontproperties=font)
axes.set_ylabel("",fontproperties=font)
axes.legend(prop=font)
plt.show()
```
